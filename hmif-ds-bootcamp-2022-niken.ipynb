1.2 Tunjukan provinsi mana yang memiliki perbandingan jumlah laki-laki terhadap perempuan paling tinggi termasuk jumlahnya 

```python
# write your code here
def ratio(a,b):
    if (a+b) == 0.0:
        return "NaN"
    else:
        return str(a/(a+b))

schools['l_per_p (decimals)']=schools.apply(lambda row: ratio(float(row.siswa_laki_laki), float(row.siswa_perempuan)), axis=1)
schools.sort_values(by='l_per_p (decimals)')
schools_filter=schools[schools['l_per_p (decimals)'] != "NaN"]
schools_filter.sort_values(by=['l_per_p (decimals)', 'siswa_laki_laki'], ascending=False)[['l_per_p (decimals)', 'siswa_laki_laki']].head(5)
```

`l_per_p (decimals)` menunjukkan perbandingan jumlah laki-laki terhadap perempuan dan `siswa_laki_laki` adalah jumlah laki-laki. Tabel di atas menunjukkan sekolah dengan perbandingan jumlah laki-laki terhadap perempuan dan jumlah laki-laki tertinggi yaitu:

```python
schools_filter.loc[[10518]]
```

dengan perbandingan jumlah laki-laki terhadap perempuan senilai **1.0** dan jumlah laki-laki tertinggi senilai **710.0**

```python
# write your code here
#Menyaring data dengan sekolah jenjang SMA
schools_SMA=schools[schools['bentuk']=="SMA"]
#Menjumlahkan jumlah setiap lab
schools_SMA[['jumlah_lab_ipa','jumlah_lab_biologi','jumlah_lab_biologi','jumlah_lab_fisika','jumlah_lab_bahasa','jumlah_lab_ips','jumlah_lab_komputer']].sum().sort_values(ascending=False)
```

1.4 Tunjukan jumlah SMA yang semua ruang kelas dan perpustakaannya memiliki kondisi baik

```python
# write your code here
schools_SMA=schools[schools['bentuk']=="SMA"]
schools_filter_ruang_perpustakaan = schools_SMA[(schools_SMA['jumlah_ruang_kelas_baik']/schools_SMA['ruang_kelas']==1.0) & (schools_SMA['perpustakaan_baik']/schools_SMA['perpustakaan']==1.0)]
schools_filter_ruang_perpustakaan[['jumlah_ruang_kelas_baik','ruang_kelas','perpustakaan_baik','perpustakaan']]
```

`schools_filter_ruang_perpustakaan` adalah data yang berisi SMA dengan semua ruang kelas dan perpustakaannya memiliki kondisi baik. Dengan demikian, jumlah SMA yang semua ruang kelas dan perpustakaannya memiliki kondisi baik adalah `len(schools_filter_ruang_perpustakaan)` yang menghasilkan total **1074**.

```python
len(schools_filter_ruang_perpustakaan)
```

1.5 Dari semua sekolah jenjang SD, provinsi mana yang memiliki kondisi perpustakaan paling baik (ditunjukkan dengan persentasi dari SD yang memiliki perpustakaan dalam kondisi baik setiap daerah )?

```python
# write your code here
schools_SD=schools[schools['bentuk']=="SD"]
jumlah_SD_perpustakaan_baik_provinsi=schools_SD[schools_SD['perpustakaan_baik']/schools_SD['perpustakaan']>0.0]["provinsi"].value_counts()
jumlah_SD_perpustakaan_provinsi=schools_SD["provinsi"].value_counts()
presentase_perpustakaan = []
provinsi = []
for i in range(0,34):
    provinsi.append(i)
    presentase_perpustakaan.append(int(jumlah_SD_perpustakaan_baik_provinsi[i])/int(jumlah_SD_perpustakaan_provinsi[i]))
presentase_perpustakaan = pd.DataFrame(
    {
        "provinsi":pd.Series(provinsi),
        "presentase_perpustakaan":pd.Series(presentase_perpustakaan)
    }
)
presentase_perpustakaan.sort_values(by='presentase_perpustakaan',ascending=False).head()
```

Berdasarkan data tersebut, provinsi dengan kondisi perpustakaan paling baik adalah provinsi dengan ID **6**. 

1.6 Tunjukkan perbandingan rata-rata akreditasi sekolah yang memiliki sumber listrik dan tidak (semakin kecil rata-rata akreditasi semakin baik akreditasinya) 

```python
# write your code here
# Rata-rata akreditasi untuk sekolah dengan sumber listrik
print("Rata-rata akreditasi untuk sekolah dengan sumber listrik : ",schools[schools['sumber_listrik']==True]['akreditasi'].mean())
print("Rata-rata akreditasi untuk sekolah tanpa sumber listrik  :",schools[schools['sumber_listrik']==False]['akreditasi'].mean())
```
